```python
#create git hash lists that contain README
def Extract_name1(repo_dir):
    os.chdir(repo_dir)
    # str = os.popen("git log --pretty=format:%H -- README.md > " + output_path).read()
    str = os.popen("git log --pretty=format:%aN -- Contributors.md").read()
    gitlist_file = open("authorName.csv", 'w')
    gitlist_file.write(str)
    gitlist_file.close()

def Extract_name2(repo_dir):
    os.chdir(repo_dir)
    # str = os.popen("git log --pretty=format:%H -- README.md > " + output_path).read()
    str = os.popen("git log --pretty=format:%aN").read()
    gitlist_file = open("authorName1.csv", 'w')
    gitlist_file.write(str)
    gitlist_file.close()
    
###SETTING PATH here
repo_dir = "/Users/ifraz-re/project_2/first-contributions/"
# gitlist_all = "/Users/pongbhop/Documents/python_workstation/npmrepo/npm_gitlist_all.txt"
# gitlist_withreadme = "/Users/pongbhop/Documents/python_workstation/Prelim/output/npm_gitlist_withreadme.txt"

Extract_name1(repo_dir)
Extract_name2(repo_dir)


```

```python
def execute_command(cmd, work_dir):
    pipe = subprocess.Popen(cmd, shell=True, cwd=work_dir, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    (out, error) = pipe.communicate()
    return out, error
    pipe.wait()
```

```python
repo_dir = "/Users/ifraz-re/project_2/first-contributions/"
git_cmd = "git log --pretty=format:%b"
log = (str(execute_command(git_cmd, repo_dir)).replace("b'",'').replace('b"','').replace('(','',1).replace("\\'","'").split("\\n"))[:-1]
pd.DataFrame(log)
```

```python
mylist = []
for l in log:
    sublist = l.split(",")
    mylist.append(sublist)
mylist
```

```python
mydf = pd.DataFrame(mylist,columns=["author_name","committer_name","committer_Email","committer_date","5","6"])
mydf = mydf[["author_name","committer_name","committer_Email","committer_date","5","6"]]
mydf
```
