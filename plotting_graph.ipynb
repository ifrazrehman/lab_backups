```python
committer_id=[]
commits_sum=[]
data = pd.read_csv("project-records/commits_projects_summary.csv")
temp1_comm = data['column_x']
temp1_sum = data['commits_sum']


for x in temp1_comm:
#     c = 'A' + str(x)
    committer_id.append(x)
    
for y in temp1_sum:
    commits_sum.append(y)
```

```python
owner_id=[]
pullreq_count=[]
data2 = pd.read_csv("project-records/pullreq_projects_summary.csv")
temp2_own= data2['column_y']
temp2_sum= data2['total_pullreq']

for i in temp2_own:
#     s = "A" + str(i)
    owner_id.append(i)

for z in temp2_sum:
    pullreq_count.append(z)
    
```

```python
owner_id2=[]
issues_count=[]
data3 = pd.read_csv("project-records/summary_issues_projects.csv")
temp3_own= data3['column_z']
temp3_sum= data3['total_issues']

for n in temp3_own:
    owner_id2.append(n)
for m in temp3_sum:
    issues_count.append(m)
```

```python
owner_id3=[]
issues_comm_count=[]
data4 = pd.read_csv("project-records/summary_issue_comments.csv")
temp4_own= data4['column_a']
temp4_sum= data4['total_issue_comments']

for a in temp4_own:
    owner_id3.append(a)
for b in temp4_sum:
    issues_comm_count.append(b)
```

```python
# plt.boxplot(committer_id, commits_sum)
# plt.ylabel("commits")
# plt.xlabel("1")
# plt.title("BOX plot for number of commits")
```

```python
trace1 = go.Box(
    y = commits_sum,
    x = committer_id
)

trace2 = go.Box(
    y = pullreq_count,
    x = owner_id
)

trace3 = go.Box(
    y = issues_count,
    x = owner_id2
)

trace4 = go.Box(
    y = issues_comm_count,
    x= owner_id3
)

data = [trace1,trace2,trace3,trace4]
layout = go.Layout(
    title = "BOX plot for number of commits(trace0) & pullrequest(trace1)"
)

init_notebook_mode(connected=True)
fig = go.Figure(data, layout=layout)
iplot(fig, show_link=False)
```
