```python
import sys, os, re
import io
from subprocess import check_output
import subprocess
import pandas as pd
import datetime
from datetime import date
```

```python
def execute_command(cmd, work_dir):
    pipe = subprocess.Popen(cmd, shell=True, cwd=work_dir, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    (out, error) = pipe.communicate()
    return out, error
    pipe.wait()
```

```python
repo_dir = "/Users/ifraz-re/project_2/first-contributions/"
git_cmd = "git log -p | grep '^[-+]' | grep -Ev '^(--- a/|\+\+\+ b/)'"
log = (str(execute_command(git_cmd, repo_dir)).replace("b'",'').replace('b"','').replace('(','',1).replace("\\'","'").split("\\n"))[:-1]
pd.DataFrame(log)
```

```python
mylist=[]
for l in log:
#     sublist=l.split("\n")
    if "(htt" in l:
        ll = l.replace("++","").replace("+-","").replace("--","").replace("+*","").replace("+[","[").replace("-[","[").replace("github/","github.com").replace("githib","github").replace("githup","github").replace("girhub","github").replace("github/com","github.com")
        lll = str(re.sub("([-+0-9]+\.)","",ll)).replace("\\t","")
        mylist.append(lll)
mylist
```
